## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
agg_amount = orders.groupby(['CustomerID']).agg({'amount_spent':sum})
agg_amount.head(10)
```

```python
P_95 = np.percentile(agg_amount['amount_spent'],95)
P_75 = np.percentile(agg_amount['amount_spent'],75)
P_75
```

```python
def customers_add(x):
    if x > P_95:
        return 'VIP Customer'
    elif x < P_95 and x >= P_75:
        return 'Preferred Customer'
    else:
        return 'Non Preferred Customer'
    
agg_amount['customer_label'] = agg_amount['amount_spent'].apply(customers_add)

agg_amount
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

country_amount = orders.groupby(['CustomerID','Country']).agg({'amount_spent':sum})
P_95 = np.percentile(agg_amount['amount_spent'],95)
P_75 = np.percentile(agg_amount['amount_spent'],75)
```

```python
def customers_vvip(x):
    if x > P_95:
        return 'VIP Customer'
    elif x < P_95 and x >= P_75:
        return 'Preferred Customer'
    else:
        return 'Non Preferred Customer'
    
country_amount['customer_label'] = country_amount['amount_spent'].apply(customers_vvip)

country_amount
```

```python
country_amount['customer_label'].value_counts()
```

```python
country_amount = country_amount.reset_index()
country_amount
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
vip_and_Preferred = country_amount[country_amount['customer_label'].isin(['VIP Customer','Preferred Customer'])]
vip_and_Preferred
```
